# Πυρηνική Αποσύνθεση

Ως *$k$-πυρήνας* ενός γράφου ορίζεται ένας υπογράφος του τέτοιος ώστε όλοι οι κόμβοι που ανήκουν σε αυτόν τον υπογράφο είναι συνδεμένοι με τουλάχιστον $k$ άλλους κόμβους του υπογράφου. Ο *πυρηνικός αριθμός* (core number) ενός κόμβου είναι ο αριθμός του $k$-πυρήνα στον οποίο ανήκει με το μέγιστο $k$.

Για να βρούμε τους $k$-πυρήνες ενός γράφου, για κάποιο συγκεκριμένο $k$, αρκεί να αφαιρέσουμε όλους τους κόμβους που έχουν λιγότερους από $k$ συνδέσμους. Οι κόμβοι που μένουν θα συνιστούν έναν ή παραπάνω γράφους, υπογράφους του αρχικού, ο καθένας από τους οποίους αποτελείται από κόμβους με τουλάχιστον $k$ συνδέσμους, άρα ο κάθε τέτοιος υπογράφος είναι $k$-πυρήνας. 

Για να βρούμε τώρα τους πυρηνικούς αριθμούς όλων των κόμβων ενός γράφου εργαζόμαστε περίπου αντίστοιχα. Στην αρχή αφαιρούμε όλους τους κόμβους που έχουν έναν σύνδεσμο. Οι κόμβοι που μένουν ανήκουν σε έναν 2-πυρήνα. Στη συνέχεια αφαιρούμε όλους τους κόμβους που έχουν ένα σύνδεσμο, αφού αφαιρέθηκαν οι προηγούμενοι. Οι κόμβοι που μένουν ανήκουν σε έναν 3-πυρήνα, κ.ο.κ. 

Μπορούμε να βρούμε τους πυρηνικούς αριθμούς των κόμβων ενός γράφου με πιο έξυπνο τρόπο. Πάλι θα βρίσκουμε και θα αφαιρούμε συγκεκριμένους κόμβους, αλλά θα χρησιμοποιήσουμε μια ουρά ελαχίστων για τον εντοπισμό του προς διαγραφή κόμβου. Επιπλέον, κάθε φορά που θα διαγράφουμε ένα κόμβο, θα ενημερώνουμε τους εν δυνάμει πυρηνικούς αριθμούς των γειτόνων του, λαμβάνοντας υπόψη τη μορφή που έχει ο γράφος μετά από κάθε διαγραφή.

Ο αλγόριθμος χρησιμοποιεί έναν σωρό ελαχίστων $\mathit{mh}$, τον οποίο δημιουργούμε στη γραμμή 1. Στη γραμμή 2 δημιουργούμε τον πίνακα $d$ στον οποίο θα αποθηκεύσουμε τον βαθμό (degree) κάθε κόμβου του γράφου. Στη γραμμή 3 δημιουργούμε τον πίνακα $p$ στον οποίο θα αποθηκεύσουμε τον εν δυνάμει (potential) πυρηνικό αριθμό κάθε κόμβου του γράφου. Στη γραμμή 4 δημιουργούμε τον πίνακα $\mathit{core}$ στον οποίο θα αποθηκεύσουμε τα αποτελέσματα του αλγορίθμου. Στο τέλος της εκτέλεσής του, το στοιχείο $\mathit{core}[i]$ θα περιέχει τον πυρηνικό αριθμό του κόμβου $i$.

Στις γραμμές 5-10 προχωράμε σε αρχικοποιήσεις τιμών. Συγκεκριμένα, αρχικοποιούμε τον πίνακα $d$ (γραμμή 6) και τον πίνακα $p$. Η πρώτη εκτίμηση που έχουμε για τον πυρηνικό αριθμό ενός κόμβου είναι να είναι ίσος με το βαθμό του (γραμμή 7). Στη χειρότερη περίπτωση ο κόμβος μπορεί να έχει πυρηνικό αριθμό ίσο με μηδέν, που είναι η τιμή που δίνουμε στον πίνακα των τελικών αποτελεσμάτων (γραμμή 8). Ο αλγόριθμος θα χρησιμοποιήσει το σωρό ελαχίστων για να βρίσκουμε κάθε φορά τον κόμβο με τον ελάχιστο εν δυνάμει πυρηνικό αριθμό. Δημιουργούμε λοιπόν έναν πίνακα δύο στοιχείων $\mathit{pn}$ (potential-node) που περιέχει τον εν δυνάμει πυρηνικό αριθμό κάθε κόμβου και τον ίδιο τον κόμβο (γραμμή 9) και εισάγουμε τον πίνακα στον σωρό ελαχίστων.

Το κύριο μέρος του αλγορίθμου είναι η επανάληψη στις γραμμές 11-20, η οποία εκτελείται όσο δεν είναι άδειος ο σωρός ελαχίστων (γραμμή 11). Ξεκινάμε βγάζοντας το ελάχιστο στοιχείο $t$ (top) από τον σωρό ελαχίστων (γραμμή 12). Το ελάχιστο στοιχείο είναι ένας πίνακας δύο στοιχείων με τον εν δυνάμει πυρηνικό αριθμό ($t[0]$) και τον αντίστοιχο κόμβο ($t[1])$. Αυτός είναι ο κόμβος με τον ελάχιστο εν δυνάμει πυρηνικό αριθμό.

Στην πρώτη επανάληψη, ο κόμβος αυτός είναι ο κόμβος με τον ελάχιστο αριθμό συνδέσμων (ή κάποιος από του κόμβους με τον ελάχιστο αριθμό συνδέσμων σε περίπτωση ισοπαλίας). Ο κόμβος αυτός δεν μπορεί να έχει πυρηνικό αριθμό μεγαλύτερο από τον αριθμό των συνδέσμων του, γιατί σε κάθε υπογράφο στον οποίο ανήκει οι υπόλοιποι κόμβοι έχουν ίσο ή μεγαλύτερο αριθμό συνδέσμων (θυμηθείτε ότι στην αρχικοποίηση, ο εν δυνάμει πυρηνικός αριθμός έγινε ίσος με το βαθμό του κόμβου). Έτσι, αποθηκεύουμε στον πίνακα των τελικών αποτελεσμάτων τον εν δυνάμει πυρηνικό αριθμό του κόμβου, αφού δεν είναι απλώς εν δυνάμει, αλλά ο πραγματικός πυρηνικός αριθμός του. Αφού το κάνουμε αυτό, ελέγχουμε αν, έχοντας αφαιρέσει ένα στοιχείο από το σωρό ελαχίστων, αυτός έχει αδειάσει (γραμμή 14). Στην περίπτωση αυτή δεν έχουμε κάτι άλλο να κάνουμε, βγαίνουμε από την επανάληψη, και επιστρέφουμε τα αποτελέσματα (γραμμή 21).

Εκτός όμως και αν ο γράφος μας έχει μόνο ένα κόμβο, δεν θα έχουμε τελειώσει, και στο σωρό ελαχίστων θα υπάρχουν οι υπόλοιποι κόμβοι. Για τους γείτονες του κόμβου τον οποίο εξεργαζόμαστε ισχύει το εξής. Είτε θα ανήκουν στον ίδιο $k$-πυρήνα με τον κόμβο που επεξεργαζόμαστε, είτε σε έναν διαφορετικό. Αν ο γείτονας ανήκει στον ίδιο $k$-πυρήνα με τον τρέχοντα κόμβο, τότε ο πυρηνικός αριθμός του θα είναι ίδιος με τον πυρηνικό αριθμό του τρέχοντα κόμβου.
Αν ανήκει σε διαφορετικό, θα πρέπει να αφαιρέσουμε από τον αριθμό συνδέσμων του γείτονα ένα, γιατί ο σύνδεσμος με τον τρέχοντα κόμβο δεν θα ανήκει στον $k$-πυρήνα και θα πρέπει να αγνοηθεί. Αυτό το κάνουμε στη γραμμή 16, με την οποία διαγράφουμε, τόπο τινά, τον κόμβο από τον γράφο· προσέξτε ότι στην πραγματικότητα δεν τον διαγράφουμε, όπως στον αρχικό αλγόριθμο, αρκεί να μην τον μετράμε πλέον στους βαθμούς των υπόλοιπων κόμβων.

Ο εν δυνάμει πυρηνικός αριθμός του γείτονα λοιπόν θα είναι ο μεγαλύτερος από τον πυρηνικό αριθμό του τρέχοντα κόμβου (αν ανήκει στον ίδιο $k$-πυρήνα με αυτόν) και τον ενημερωμένο βαθμό του γείτονα (αν δεν ανήκει στον ίδιο $k$-πυρήνα). Πριν προχωρήσουμε στην ενημέρωση του πυρηνικού αριθμού, κρατάμε σε μία μεταβλητή $\mathit{opn}$ (old potential node) το υπάρχον στοιχείο του σωρού ελαχίστων που αντιστοιχεί στον γείτονα. Κατόπιν προχωρούμε στην ενημέρωση του πυρηνικού αριθμό σύμφωνα με τα όσα είπαμε, στη γραμμή 18. Αφού το κάνουμε αυτό, πρέπει να ενημερώσουμε την ουρά ελαχίστων. Δημιουργούμε τον νέο, ενημερωμένο κόμβο $\mathit{npn}$ (new potential node, γραμμή 19), και ενημερώνουμε την ουρά, ζητώντας να αλλαχτεί ο παλιός, υπάρχον κόμβος, με τον νέο (γραμμή 20).

Με τον τρόπο αυτό, στο τέλος κάθε επανάληψης η ουρά ελαχίστων θα μπορεί να μας δίνει, στην επόμενη επανάληψη, τον κόμβο με τον ελάχιστο εν δυνάμει πυρηνικό αριθμό. Σκεφτόμενοι όπως και πριν, μπορούμε να δούμε ότι αυτός θα είναι και ο πραγματικός πυρηνικός αριθμός, γιατί αφού αφαιρούμε, έστω νοητικά, κόμβους από το γράφο, κάθε επανάληψη δρα σαν να ξεκινούσαμε από την αρχή σε ένα νέο γράφο, με έναν λιγότερο κόμβο, αρχικοποιημένο όπως πρέπει.

## Απαιτήσεις Προγράμματος

Κάθε φοιτητής θα εργαστεί σε αποθετήριο στο GitHub. Για να αξιολογηθεί μια εργασία θα πρέπει να πληροί τις παρακάτω προϋποθέσεις:

* Για την υποβολή της εργασίας θα χρησιμοποιηθεί το ιδιωτικό αποθετήριο του φοιτητή που δημιουργήθηκε για τις ανάγκες του μαθήματος και του έχει αποδωθεί. Το αποθετήριο αυτό έχει όνομα του τύπου `username-algo-assignments`, όπου `username` είναι το όνομα του φοιτητή στο GitHub. Για παράδειγμα, το σχετικό αποθετήριο του διδάσκοντα θα ονομαζόταν `louridas-algo-assignments` και θα ήταν προσβάσιμο στο https://github.com/dmst-algorithms-course/louridas-algo-assignments. *Τυχόν άλλα αποθετήρια απλώς θα αγνοηθούν*.

* Μέσα στο αποθετήριο αυτό θα πρέπει να δημιουργηθεί ένας κατάλογος `assignment-2019-1`.

* Μέσα στον παραπάνω κατάλογο το πρόγραμμα θα πρέπει να αποθηκευτεί με το όνομα `k_cores.py`.

* Δεν επιτρέπεται η χρήση έτοιμων βιβλιοθηκών γράφων ή τυχόν έτοιμων υλοποιήσεων των αλγορίθμων, ή τμημάτων αυτών. Επιτρέπεται η χρήση δομών δεδομένων της Python όπως στοίβες, λεξικά, κ.λπ.

* Το πρόγραμμα θα πρέπει να είναι γραμμένο σε Python 3.

* Το πρόγραμμα θα το καλεί ως χρήστης ως εξής (όπου `python` η κατάλληλη εντολή στο εκάστοτε σύστημα):

```
python k_cores.py <input_file>
```

Στην εντολή αυτή, η παράμετρος `<input_file>` είναι ένα όνομα αρχείου που περιέχει το γράφο που θα εξετάσουμε.

Τα αρχεία των γράφων θα έχουν την εξής μορφή:
```
0 1
0 2
1 3
2 3
2 6
2 7
3 6
3 7
3 8
4 5
5 6
5 10
6 7
7 8
7 11
8 9
```
Βλέπουμε δηλαδή ότι αποτελούνται από γραμμές, όπου η κάθε γραμμή δηλώνει τη σύνδεση δύο κόμβων. Οι γράφοι θεωρούνται ότι είναι μη κατευθυνόμενοι.

Η έξοδος του προγράμματος θα είναι μία γραμμή ανά κόμβο και θα εμφανίζει τον κόμβο και τον πυρηνικό αριθμό του. Οι κόμβοι θα είναι ταξινομένοι σε αύξουσα σειρά.
